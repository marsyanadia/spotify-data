## Calling Attributes

```python
#mendefinisikan feature
def get_features(tr_id):
    meta = sp.track(tr_id)
    features = sp.audio_features(tr_id)
    #atribut yang ingin diambil
    name = meta['name']
    album = meta['album']['name']
    artist = meta['album']['artists'][0]['name'] #mengambil atribut dengan format json
    release_date = meta['album']['release_date']
    length = meta['duration_ms']
    popularity = meta['popularity']
    
    acousticness = features[0]['acousticness'] #array ke-0 (track pertama)
    danceability = features[0]['danceability']
    energy = features[0]['energy']
    instrumentalness = features[0]['instrumentalness']
    liveness = features[0]['liveness']
    loudness = features[0]['loudness']
    speechiness = features[0]['speechiness']
    tempo = features[0]['tempo']
    time_signature = features[0]['time_signature']
    
    track = [name, album, artist, release_date, length, popularity, 
             acousticness, danceability, energy, instrumentalness, liveness, 
             loudness, speechiness, tempo, time_signature]
    return track
```

## Get Data from Spotify

```python
import pandas as pd 
```

```python
def get_songs():
    tracks = []
    for i in range(0, 1000, 50):
        song = sp.search(q='year:2023', type='track', limit='50')
        for j in range(50):
            tr_id = song['tracks']['items'][j]['id'] # memperbaiki variabel yang digunakan sebagai key
            track = get_features(tr_id)
            tracks.append(track)
    df = pd.DataFrame(tracks, columns=['name', 'album', 'artist', 'release_date', 'length', 
                                       'popularity', 'acousticness', 'danceability', 'energy', 'instrumentalness', 
                                       'liveness', 'loudness', 'speechiness', 'tempo', 'time_signature'])
    return df
```
